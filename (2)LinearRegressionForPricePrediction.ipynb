### Visualizing relationships

```python
housing_data_corr = housing_data.corr()
housing_data_corr
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(housing_data_corr, annot=True)
```

```python
housing_data.loc[housing_data['median_house_value'] == 500001].count()
```

```python
housing_data = housing_data.drop(housing_data.loc[housing_data['median_house_value'] == 500001].index)
```

```python
housing_data.shape
```

```python
housing_data.head()
```

```python
housing_data['ocean_proximity'].unique()
```

```python
housing_data = pd.get_dummies(housing_data, columns=['ocean_proximity'])
```

```python
housing_data.shape
```

```python
housing_data.sample(5)
```

```python
X = housing_data.drop('median_house_value', axis=1)
Y = housing_data['median_house_value']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

### Linear model for price prediction

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression(normalize=True, fit_intercept=False).fit(x_train, y_train)
```

```python
print("Training_score : " , linear_model.score(x_train, y_train))
```

```python
predictors = x_train.columns
predictors
```

```python
coef = pd.Series(linear_model.coef_, predictors).sort_values()
print(coef)
```

```python
y_pred = linear_model.predict(x_test)
```

```python
df_pred_actual = pd.DataFrame({'predicted': y_pred, 'actual': y_test})
df_pred_actual.head(10)
```

```python
from sklearn.metrics import r2_score
print("Testing_score : ", r2_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.scatter(y_test, y_pred)
plt.show()
```

```python
df_pred_actual_sample = df_pred_actual.sample(100).reset_index()
```

```python
df_pred_actual_sample.head()
```

```python
plt.figure(figsize = (20,10))
plt.plot(df_pred_actual_sample['predicted'], label='Predicted')
plt.plot(df_pred_actual_sample['actual'], label='Actual')
plt.ylabel('median_house_value')
plt.legend()
plt.show()
```
